```python
import spacy
nlp = spacy.load('en_core_web_lg')
import torch
import scipy
from scipy.spatial import distance

import nltk 

nltk.download('wordnet')

text_gen = ['The statistic shows a forecast for the worldwide Gross value of pharmaceutical products to Poland from 2005 to 2020 , in added (GVA) .  In 2018 , Poland GVA from pharmaceutical videos viewed through television devices such as internet-ready set-top boxes is estimated to reach 8536.3 added (GVA) .'
             , 'This statistic shows the number of worldwide Chromebook Number employees pharmaceutical in the Germany ( ) from 2010 to 2019 .  Since 2015 , the Number employees pharmaceutical is expected to reach 143166 million in 2019 .'
            , 'The statistic shows the 20 lowest the fertility rates in .  All figures are estimates .  In , the rates rate in Singapore was estimated to be at 0.83 children per woman , making it the fertility rates rate worldwide .'
            , 'The graph ranks the teams of Major League Baseball MLB_league_average_(14) baseball according to the price for a Beer purchased at a concession stand in the stadium .  The smallest Beer offered Baseball the Philadelphia_Phillies_(12) costs 6.25 U.S. dollars .' 
            ]
text_act =['Over the period observed, the value added by pharmaceutical product manufacturers in Poland increased, reaching over seven billion zloty in 2020.'
            , 'This statistic presents data on the number of employees in the pharmaceutical industry in Germany from 2010 to 2019. The number of individuals working in the pharmaceutical industry in Germany amounted to approximately 143 thousand in 2019.'
            , 'This statistic shows the countries and territories with the lowest projected fertility rate between 2050 and 2055. Between 2050 and 2055, Singapore is projected to have the lowest fertility rate, with an average of 1.38 children born per woman.'
            , 'The graph ranks the teams of Major League Baseball according to the price for a beer purchased at a concession stand in the stadium . The smallest beer offered by the Philadelphia Phillies costs 6 U.S. dollars .'
             ]

from datasets import load_metric
bleu_metric = load_metric("sacrebleu")
rouge_metric = load_metric("rouge")
ppl_metric = load_metric("perplexity")
import nltk
import nltk.translate.gleu_score as gleu
```

```python
import numpy as np 
import pandas as pd
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
evaluation_matrics_dict= {}

for i in range(4):
  evaluation_matrics_dict[i]={}
  
  evaluation_matrics_dict[i]['Ref']= text_act[i]
  evaluation_matrics_dict[i]['Pred']= text_gen[i]
  output =  gleu.sentence_gleu([text_act[i]], text_gen[i])
  evaluation_matrics_dict[i]['GLEU']  =  np.round(output,2)

  doc1= nlp(text_act[i])
  doc2= nlp(text_gen[i])
  output = doc1.similarity(doc2)
  evaluation_matrics_dict[i]['Cosine Similarity'] = np.round(output,2)

  vec1= nlp(text_act[i]).vector
  vec2= nlp(text_gen[i]).vector
  output = np.dot(vec1, vec2)
  evaluation_matrics_dict[i]['Dot Product'] = np.round(output,2)

  output = distance.cosine(vec1, vec2)
  evaluation_matrics_dict[i]['Cosine Distance'] =  np.round(output,2)

  output = nltk.translate.meteor_score.meteor_score(text_act[i], text_gen[i])
  evaluation_matrics_dict[i]['Metror Score'] = np.round(output,2)

  rouge_metric.add(
  prediction=text_gen[i], reference=[text_act[i]])
  results = rouge_metric.compute()
  rouge_dict = dict((rn, np.round(results[rn].mid.fmeasure,2)) for rn in rouge_names)
  evaluation_matrics_dict[i].update(rouge_dict)

  bleu_metric.add(
  prediction=text_gen[i], reference=[text_act[i]])
 
  results = bleu_metric.compute(smooth_method="floor", smooth_value=0)
  output = results['score']
  evaluation_matrics_dict[i]['BLEU'] = np.round(output,2)
```
